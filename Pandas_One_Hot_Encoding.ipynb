```python
import pandas as pd
```

```python
df = pd.read_csv("telecom_customer_churn.csv")
```

```python
df_pd_encoded = pd.get_dummies(
    df,
    columns=[
        'Gender',
        'Married',
        'Offer',
        'Phone Service',
        'Multiple Lines',
        'Internet Service',
        'Internet Type',
        'Online Security',
        'Online Backup',
        'Device Protection Plan',
        'Premium Tech Support',
        'Streaming TV',
        'Streaming Movies',
        'Streaming Music',
        'Unlimited Data',
        'Contract',
        'Paperless Billing',
        'Payment Method'
    ],
    drop_first=True,
    dtype=int
)
df_pd_encoded
```

```python
from google.colab import files

df_pd_encoded.to_csv("telecom_churn_encoded.csv", index=False)
files.download("telecom_churn_encoded.csv")
```
