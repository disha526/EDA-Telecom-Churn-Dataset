```python
import pandas as pd

df = pd.read_csv('/content/main-telecom_customer_churn.csv')
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

num_cols = [
    'Age',
    'Tenure in Months',
    'Monthly Charge',
    'Total Charges',
    'Total Revenue'
]

target = 'Customer Status'

for col in num_cols:

    plt.figure(figsize=(8,5))

    sns.boxplot(
        x=target,
        y=col,
        data=df
    )

    plt.title(f'{col} vs {target}')
    plt.show()
```

```python
cat_cols = [
    'Contract',
    'Internet Service',
    'Internet Type',
    'Payment Method',
    'Gender',
    'Married',
    'Paperless Billing'
]

for col in cat_cols:

    cross_tab = pd.crosstab(
        df[col],
        df['Customer Status'],
        normalize='index'
    ) * 100

    print("\n")
    print("="*60)
    print(col)
    print("="*60)

    print(cross_tab)

    cross_tab.plot(
        kind='bar',
        stacked=True,
        figsize=(8,5)
    )

    plt.title(f'{col} vs Customer Status')
    plt.ylabel('Percentage')
    plt.show()
```

```python
num_cols = df.select_dtypes(include='number')

corr_matrix = num_cols.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(
    x='Monthly Charge',
    y='Total Charges',
    data=df
)

plt.title('Monthly Charge vs Total Charges')
plt.show()
```

```python
sns.scatterplot(
    x='Tenure in Months',
    y='Total Revenue',
    data=df
)

plt.title('Tenure vs Total Revenue')
plt.show()
```
