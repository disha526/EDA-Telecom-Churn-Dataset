```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("main-telecom_customer_churn.csv")
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
print("="*60)
print("NUMERICAL COLUMNS")
print("="*60)
print(numerical_cols)

print("\n" + "="*60)
print("CATEGORICAL COLUMNS")
print("="*60)
print(categorical_cols)
```

```python
for col in numerical_cols:

    print("\n" + "="*70)
    print(f"UNIVARIATE ANALYSIS : {col}")
    print("="*70)

    stats = {
        'Count': df[col].count(),
        'Missing Values': df[col].isnull().sum(),
        'Mean': df[col].mean(),
        'Median': df[col].median(),
        'Mode': df[col].mode()[0] if not df[col].mode().empty else np.nan,
        'Min': df[col].min(),
        'Max': df[col].max(),
        'Std Dev': df[col].std(),
        'Variance': df[col].var(),
        'Skewness': df[col].skew(),
        'Kurtosis': df[col].kurt()
    }

    for k, v in stats.items():
        print(f"{k:<20}: {v}")

    fig, ax = plt.subplots(1, 2, figsize=(12,4))

    sns.histplot(df[col], kde=True, ax=ax[0])
    ax[0].set_title(f'Histogram - {col}')

    sns.boxplot(x=df[col], ax=ax[1])
    ax[1].set_title(f'Boxplot - {col}')

    plt.tight_layout()
    plt.show()
```

```python

for col in categorical_cols:

    print("\n" + "="*70)
    print(f"UNIVARIATE ANALYSIS : {col}")
    print("="*70)

    print(f"Missing Values : {df[col].isnull().sum()}")
    print(f"Unique Values  : {df[col].nunique()}")

    freq_table = pd.DataFrame({
        'Count': df[col].value_counts(),
        'Percentage': round(df[col].value_counts(normalize=True)*100,2)
    })

    print(freq_table)

    plt.figure(figsize=(10,5))

    if df[col].nunique() <= 15:
        sns.countplot(y=col, data=df,
                      order=df[col].value_counts().index)

        plt.title(f'Frequency Distribution - {col}')
        plt.show()

    else:
        print(f"{col} has high cardinality ({df[col].nunique()} categories). Chart skipped.")
```
